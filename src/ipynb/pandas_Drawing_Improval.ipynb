```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load iris data
url = 'https://raw.github.com/pydata/pandas/master/pandas/tests/data/iris.csv'
iris = pd.read_csv(url)
iris.head(5)
```

```python
np.corrcoef(np.array(iris.SepalLength), np.array(iris.SepalWidth))
# iris.SepalLength.to_xarray
```

```python
iris.corr()
```

```python
# scatter_matrix
from pandas.plotting import scatter_matrix
scatter_matrix(iris, figsize=(8, 8), diagonal='kde')
```

```python
iris.boxplot()
```

```python
# 通过by参数可以计算不同分组情况下，各个字段的箱图
iris.boxplot(by='Name', figsize=(8, 8))
```

```python
# 概率密度分布
iris.plot(kind='kde')
```
