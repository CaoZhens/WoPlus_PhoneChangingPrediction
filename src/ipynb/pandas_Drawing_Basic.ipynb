```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load Data
url = 'http://s3.amazonaws.com/assets.datacamp.com/course/dasi/present.txt'
df = pd.read_table(url, sep=' ')
df = df.set_index('year')
# attentionï¼šset_index default inplace = False
print df.shape
print df.columns
```

```python
# plot method on Series
# df['boys']
df['boys'].plot()
plt.legend(loc='best')
```

```python
# plot method on DataFrame
df.plot()
```

```python
# comment: dataframe's plot is the same with n Series's plot
df.boys.plot(color='g')
df.girls.plot(color='b')
plt.legend(loc='best')
```

```python
# default kind = 'line'
# you can use other kind parameters
df.iloc[0:10,:].plot(kind='bar')
```

```python
df.iloc[0:10,:].plot(kind='barh')
```

```python
df.iloc[0:10,:].plot(kind='bar', stacked=True)
```

```python
# Curmulated bar
# df.sum(axis=1)
df.div(df.sum(axis=1), axis=0)[:10]
```

```python
# Scatter
df.plot(x='boys', y='girls', kind='scatter')
```
